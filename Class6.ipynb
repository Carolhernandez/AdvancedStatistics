```python
%matplotlib inline
import numpy as np
import scipy.stats as stats
from IPython.core.pylabtools import figsize
import matplotlib.pyplot as plt

figsize(12.5,3)
colors = ["green", "red", "grey", "orange"]

x = np.linspace(0,1)
y1, y2 = stats.beta.pdf(x, 1,1), stats.beta.pdf(x, 10,10)

p = plt.plot(x, y1, label='An objective prior \n(uninformative, \n"Principle of Indifference")')

p = plt.plot(x,y2, label = "A subjective prior \n(informative)")

p = plt.plot(x[25:], 2*np.ones(25), label = "another subjective prior")

plt.ylim(0,4)
leg = plt.legend(loc = "upper left")
plt.title("Objective vs. subjective priors (for an unknown probability)")
```

# $\Gamma$ Distribution

```python
figsize(12.5, 5)

params = [(2, 0.6)]

x = np.linspace(0.01, .99, 100)

beta = stats.gamma
for a, b in params:
    y = beta.pdf(x, a, b)
    lines = plt.plot(x, y, label = "(%.1f,%.1f)"%(a,b), lw = 3)
    plt.autoscale(tight=True)
plt.ylim(0)
```

# $\beta$ Distribution

```python
figsize(12.5, 5)

params = [(20, 4)]

x = np.linspace(0.01, .99, 100)

beta = stats.beta
for a, b in params:
    y = beta.pdf(x, a, b)
    lines = plt.plot(x, y, label = "(%.1f,%.1f)"%(a,b), lw = 3)
    plt.autoscale(tight=True)
plt.ylim(0)
```
